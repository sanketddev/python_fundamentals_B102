```python
# List Indexing
# Indexing is used to access, modify, or delete elements from a list.
# Python uses zero-based indexing, which means the first element starts at index 0
```

```python
aws_services = ["EC2", "S3", "IAM", "CloudWatch", "Lambda"]

print(aws_services)
```

```python
# req: i want to access ec2 service

print(aws_services[0])
```

```python
# req: Print a Custom Message : Let's launch an EC2 instance today!

print(f"Let's launch an {aws_services[0]} instance today!")
```

```python
# req: add VPC to the AWS services list.

aws_services.append("VPC")

print(aws_services)
```

```python
# req: add EKS to the AWS services list.

aws_services.append("EKS")

print(aws_services)
```

```python
# req: add Route 53 at index 2

aws_services.insert(2, "Route 53")

print(aws_services)
```

```python
# req: replace Lambda with ECS.

aws_services[5] = "ECS"

print(aws_services)
```

```python
# req: remove S3 permanently

del aws_services[1]

print(aws_services)
```

```python
# pop() is used to temporarily delete an element from a list.
# It creates a copy (returns the deleted item), so you can use it later in your program.
# In general, pop() deletes the last item from the list.
```

```python
# req: I want to temporarily delete the AWS service.

aws_services = ['EC2', 'Route 53', 'IAM', 'CloudWatch', 'ECS', 'VPC', 'EKS']

removed_service = aws_services.pop()

print(remove_service)
print(aws_services)
```

```python
# append() vs insert()

# append(): Adds an item at the end of the list.
# insert(): Adds an item at a specific index.


# pop() vs del()

# pop(): Temporarily deletes an item from the list and returns it for later use.
# del(): Permanently deletes an item from the list.
```
